```python
import statsmodels.api as sm
```

```python
X=data.drop(['cnt'],axis=1)
y=data['cnt']
```

```python
x=sm.add_constant(X)
x
```

```python
model=sm.OLS(y,x)
model=model.fit()
model.summary()8
```

```python
df=X.drop(['holiday','weekday','workingday','weathersit','temp','atemp','hum','windspeed'],axis=1)
df.head()
```

```python
fitted_vals = model.predict()
resids = model.resid

sn.regplot(x=fitted_vals, y=resids)
plt.title('Residuals vs. Predicted Value')
```

```python
sn.distplot(model.resid)
```

```python
resids_standardized = model.get_influence().resid_studentized_internal
```

```python
sn.regplot(x=fitted_vals, y=np.sqrt(np.abs(resids_standardized)),line_kws={'color':'orange'})
```
